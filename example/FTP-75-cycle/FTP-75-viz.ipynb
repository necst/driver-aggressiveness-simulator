```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('FTP-75.csv')
df.drop(index=0, inplace=True, axis=0)
# df.drop('Phase', inplace=True, axis=1)
df = df.dropna(axis='index')
```

```python
df
```

```python
xt = np.array(df['Time'], dtype=float)
yv = np.array(df['Set Spd'], dtype=float)
yp = np.array(df['Phase'], dtype=float)
```

```python
plt.figure(figsize=(20,8))
plt.ylabel("Desired Vehicle Speed [mph]")
plt.xlabel("Time [s]")
plt.plot(xt, yv, label='Desired Vehicle Speed')
plt.plot(xt, 10*yp, label='Phase')
plt.legend()
plt.show()
```
